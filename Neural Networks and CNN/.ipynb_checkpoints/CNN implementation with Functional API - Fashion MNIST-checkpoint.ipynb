# Image Classification

## 1. Data Collection

```python
# loading the dataset
from tensorflow.keras.datasets import fashion_mnist
(xtrain,ytrain),(xtest,ytest) = fashion_mnist.load_data()
```

```python
print(xtrain.shape)
print(ytrain.shape)
print(xtest.shape)
print(ytest.shape)
```

```python
class_names = ["T-shirt/top","Trouser","Pullover","Dress","Coat","Sandal",
               "Shirt","Sneaker","Bag","Ankle boot"]
len(class_names)
```

# 2. Data Exploration

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 3. Data Preprocessing

```python
print(xtrain.shape)
print(xtest.shape)
```

```python
# reshape images to invlove the channel value
xtrain = xtrain.reshape(60000,28,28,1)
xtest = xtest.reshape(10000,28,28,1)
print(xtrain.shape)
print(xtest.shape)
```

```python
# scale the images
xtrain = xtrain/255
xtest = xtest/255
```

```python
print(ytrain.shape)
print(ytest.shape)
ytrain
```

```python
# one hot encode the labels
from tensorflow.keras.utils import to_categorical
ytrain = to_categorical(ytrain)
ytest = to_categorical(ytest)
print(ytrain.shape)
print(ytest.shape)
```

```python
ytrain[1]
```

# 4. Modelling the network

<img src = "lenet5.PNG">

```python
from tensorflow.keras import models,layers
import tensorflow as tf
```

```python
tf.random.set_seed(5)
# creating a sequential model
model = models.Sequential()

# add a convlutional layer - in the very first layer of seq model, input_shape has to be used
model.add(layers.Conv2D(filters=6,kernel_size=(3,3),activation='relu',input_shape=(28,28,1)))
# add a convolutional layer
model.add(layers.Conv2D(filters=10,kernel_size=(3,3),activation='relu'))
# add a maxpooling layer
model.add(layers.MaxPooling2D(pool_size=(2,2)))


# add a convolutional layer
model.add(layers.Conv2D(filters=20,kernel_size=(3,3),activation='relu'))
# add a convolutional layer
model.add(layers.Conv2D(filters=30,kernel_size=(3,3),activation='relu'))
# add a maxpooling layer
model.add(layers.MaxPooling2D(pool_size=(2,2)))

# =================== Flatten =========================== #
model.add(layers.Flatten())
# ================= Classification ====================== #
model.add(layers.Dense(120,activation='relu'))
model.add(layers.Dense(84,activation='relu'))
model.add(layers.Dense(10,activation='softmax'))
```

```python
from tensorflow.keras.utils import plot_model
import pydot,graphviz
plot_model(model)
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
model.fit(xtrain,ytrain,epochs=10,batch_size=1000,verbose=True,validation_data=(xtest,ytest))
```

```python
model.fit(xtrain,ytrain,epochs=10,batch_size=1000,verbose=True,validation_data=(xtest,ytest))
```

# 5. Performance Analysis

```python
ypred = model.predict_classes(xtest)
ytest2 = np.array([np.argmax(i) for i in ytest])
```

```python
ypred
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest2,ypred))
```
