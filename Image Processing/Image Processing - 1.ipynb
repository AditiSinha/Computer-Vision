# Images and Image Processing

- Images are composed of pixels
- all the computers, software are programmed to process pixels in 8 bit size
- the pixels values can range from 0000 0000 to 1111 1111 i.e. from 0 to 255

There are primarily two types of images
- Grayscale images - each pixel has one value raning from 0 to 255
- Color image - each pixel has 3 values ranging from 0 to 255, each value is for - R, G, B

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([[0,255,0,255],
             [255,0,255,0],
             [0,255,0,255],
             [255,0,255,0]])
plt.imshow(x,cmap='gray')
plt.show()
```

```python
x = np.array([[0,30,80,120],
             [255,220,200,160],
             [0,255,0,255],
             [255,0,255,0]])
plt.imshow(x,cmap='gray')
plt.show()
```

### Color images
- each pixel consists of 3 values
   - [0,0,0] = Black color
   - [255,0,0] = Red Color
   - [0,255,0] = Green Color
   - [0,0,255] = Blue Color
   -  [255,255,255] = White color

```python
y = np.array([[[0,0,0],[0,255,0]],
              [[0,0,255],[255,0,0]],
              [[255,255,255],[255,255,0]]])

plt.imshow(y)
plt.show()
```

### Morphological Analysis
- erosion  - loss of pixels from edges of the objects
- dilatoin - gain of pixels to the edges of the objects

```python
img = cv2.imread(r"E:\image_data\np2.jpg",0)
img.shape
```

```python
plt.imshow(img,cmap='gray')
plt.show()
```

```python
kernel = np.ones((7,7))
img2 = cv2.erode(img,kernel)
plt.imshow(img2,cmap='gray')
plt.show()
```

```python
#dilatio
img3 = cv2.dilate(img2,kernel)
plt.imshow(img3,cmap='gray')
plt.show()
```

```python
# Face detection
fd = cv2.CascadeClassifier(r"E:\image_data\haarcascade_frontalface_alt.xml")
img = plt.imread(r"E:\image_data\dhoni-kohli.jpg")
corners = fd.detectMultiScale(img)
corners
```

```python
for x,y,w,h in corners:
    cv2.rectangle(img,(x,y),(x+w,y+h),[255,0,0],2)
plt.imshow(img)
plt.show()
```

```python
# working with videos
# creating a video object
vid = cv2.VideoCapture(r"F:\videos\video3.mp4")
while True:
    flag,img = vid.read()
    if flag:
        corners = fd.detectMultiScale(img)
        for x,y,w,h in corners:
            cv2.rectangle(img,(x,y),(x+w,y+h),[0,0,255],2)
        cv2.imshow("image",img)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break
    else:
        break
vid.release()
cv2.destroyAllWindows()        
```
